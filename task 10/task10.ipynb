### Data Pre-Processing

### Manage Null Values

```python

df = df.drop('Max HR', axis=1)
```

```python

df = df.drop('Sex', axis=1)
```

```python

def fillnaObjectMode(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])

columns = ['Age', 'Cholesterol', 'BP']
fillnaObjectMode(columns)
```

```python

def fillnaIntMode(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])

columns = ['Age','Cholesterol','BP','Heart Disease','Chest pain type','Slope of ST','ST depression']
fillnaIntMode(columns)
```

```python
def fillnaFloatMean(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mean())

columns = ['Age','BP','Chest pain type']
fillnaFloatMean(columns)
```

```python
df.info()
```

Manage Data Types

```python
df['Age'] = df['Age'].astype('int64')
```

```python

def changeFloatToInt(cols):
    for i in cols:
        df[i] = df[i].astype('int64')

columns = ['Age','BP','Chest pain type']
changeFloatToInt(columns)
```

```python
df.info()
```

```python
def convertObjToInt(cols):
    for i in cols:
        data = pd.DataFrame({i:df[i].unique()})
        data_LE = LabelEncoder()
        data_LE.fit(np.ravel(data))    
        df[i] = data_LE.transform(df[i])
        
columns = ['Age', 'BP', 'Chest pain type']
convertObjToInt(columns)
```
