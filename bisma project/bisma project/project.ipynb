### Data Pre-Processing

### Manage Null Values

```python

print(df.isnull().sum())


missing_percent = (df.isnull().sum() / len(df)) * 100
print(missing_percent)
```

```python
df = df.drop('Income', axis=1)
```

```python
def fillnaObjectMode(cols, df):
 
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])
        print(f"Filled missing values in {i} with mode {df[i].mode()[0]}")
columns = ['Unusual_Bleeding', 'Menses_score']
fillnaObjectMode(columns, df)

```

```python

def fillnaIntMode(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])

columns = ['number_of_peak', 'Age', 'Length_of_cycle', 'Estimated_day_of_ovulution',
           'Length_of_Leutal_Phase', 'Length_of_menses', 'Weight', 'BMI', 'Mean_of_length_of_cycle']
fillnaIntMode(columns)
```

```python
def fillnaFloatMean(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mean())

columns = ['Weight', 'BMI', 'Mean_of_length_of_cycle']
fillnaFloatMean(columns)
```

```python
df.info()
```

### Manage Data Types

```python
#converted to integers
df['number_of_peak'] = df['number_of_peak'].astype('int64')
df['Age'] = df['Age'].astype('int64')
df['Length_of_cycle'] = df['Length_of_cycle'].astype('int64')
df['Estimated_day_of_ovulution'] = df['Estimated_day_of_ovulution'].astype('int64')
df['Length_of_Leutal_Phase'] = df['Length_of_Leutal_Phase'].astype('int64')
df['Length_of_menses'] = df['Length_of_menses'].astype('int64')
df['Weight'] = df['Weight'].astype('int64')
df['Mean_of_length_of_cycle'] = df['Mean_of_length_of_cycle'].astype('int64')
df['Menses_score'] = df['Menses_score'].astype('int64')
```

```python
df['BMI'] = df['BMI'].astype('float64')
```

```python
df['Unusual_Bleeding'] = df['Unusual_Bleeding'].astype('object')
```

```python
def changeFloatToInt(cols, df):

    for col in cols:
        df[col] = df[col].fillna(0)  
        df[col] = df[col].astype('int64')
        print(f"Converted {col} to int64")

float_cols_to_int = [ 'BMI', 'Weight','Menses_score'] 
changeFloatToInt(float_cols_to_int, df)
```

```python
def convertObjToInt(cols):
    for i in cols:
        data = pd.DataFrame({i:df[i].unique()})
        data_LE = LabelEncoder()
        data_LE.fit(np.ravel(data))    
        df[i] = data_LE.transform(df[i])
        
columns = ['Unusual_Bleeding','Height']
convertObjToInt(columns)
```

### Train - Test Splitting

```python
import pandas as pd
df = pd.read_csv('period-Copy.csv')
df.columns
```

```python
X = df.drop('Unusual_Bleeding', axis=1)   
y = df['Unusual_Bleeding']  
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X,y, test_size=0.3, shuffle=False) 
```

```python
print(train_X.shape, train_y.shape)
print(test_X.shape, test_y.shape)
```

### Applying Classifier / Model Training

```python
from sklearn.svm import SVC

model_svc = SVC()
model_svc.fit(train_X, train_y)

print(model_svc)
```

```python
model_pred_svc = model_svc.predict(test_X)
model_pred_svc
```

```python
from sklearn.metrics import accuracy_score
model_acc_svc = accuracy_score(test_y, model_pred_svc)
print(round(model_acc_svc,3)*100, "%")
```
